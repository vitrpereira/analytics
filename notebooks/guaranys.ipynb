### *__Scrubing__*

```python
#Limpando as colunas
moradia.columns = [col.split(". ", 1)[1] if ". " in col else col for col in moradia.columns]

list(enumerate(moradia))[:10]
```

```python
#Trocando os valores binários por categóricos (1, 0 -> Sim, Nao)
cols_to_filter = ['Do total de cômodos, quantos são exclusivamente dormitórios?', 'Quantos cômodos há na casa?']
moradia_filtered = moradia.drop(columns=cols_to_filter)

colunas_binarias = moradia_filtered.select_dtypes(include=['int64'])

for colunas, valores in colunas_binarias.items():
    moradia[colunas] = valores.replace({0: 'Não', 1: 'Sim'})

moradia.head()
```

### *__Exploring__*

```python
def observacao(df, pergunta, objeto_para_renomear=None, novo_nome=None, plot_grafico=bool, tipo_grafico='bar', *args, **kwargs):
    """ 
    Esta função facilita o processo de análise.

    df: deve ser um pd.DataFrame
    coluna_pergunta: a pergunta a ser respondida. É tirada das colunas do questionário, 
    objeto_para_renomear: caso exista algum objeto dentro da coluna a ser renomeada, passe em integralidade ou como um raw string. Defualt é None
    novo_nome: o novo nome do objeto renomeado. deve ser usado em conjunto com o argumento acima. Default é None
    plot_grafico: fornece a opção de plotar um gráfico junto com a observação 
    tipo_grafico: permite escolher o tipo de gráfico, caso a opção acima seja verdadeira
    """
    problema = (
        df[pergunta]
        .value_counts()
        .to_frame()
        .rename({f'{objeto_para_renomear}': f'{novo_nome}'})
        )

    problema['percentual'] = round((df[pergunta].value_counts(normalize=True))*100, 2)
    #problema = problema.set_index(pergunta)

    if plot_grafico == True:
        plot = problema[pergunta].plot(kind=tipo_grafico, figsize=(9,6), 
                                                color='#61d199', rot=45, 
                                                title=pergunta)
    else:
        print(" ")
    
    plt.show()

    return problema

print(observacao.__doc__)
```

```python
#Qual o material predominante no piso da casa
material_piso = observacao(
    df=moradia,
    pergunta='Qual é o material predominante no piso interno da casa?',
    # objeto_para_renomear=r'Sucata sobre a terra (retalhos de plástico, lona, metal, papelão, etc)',
    # novo_nome='Sucata',
    plot_grafico=False,
    tipo_grafico='line')

material_piso
```
